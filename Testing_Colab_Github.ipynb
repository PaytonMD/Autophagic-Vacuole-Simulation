```python
import numpy as  np
import matplotlib.pyplot as plt
```

```python
# The newer, more-preferred, object-oriented / API style,
# rather than using the older pyplot interface;
# see https://matplotlib.org/stable/tutorials/introductory/lifecycle.html
# Code from https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Create just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```
